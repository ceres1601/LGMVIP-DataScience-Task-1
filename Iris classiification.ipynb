```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mtp
import warnings
```

```python
column=["sepal lemgth", "sepal width", "petal length", "petal width", "class label"]
```

```python
df=pd.read_csv("bezdekIris.data",names = column)
```

```python
x=df.drop("class label",axis =1)
y=df["class label"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.metrics import accuracy_score
```

```python
# Traning the model using SVC
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train,y_train)
```

```python
#Checking accuracy
pred1=svc.predict(x_test)
acc1=accuracy_score(pred1,y_test)
acc1
```

```python
#Traning the model using logistic regresson
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(x_train,y_train)
```

```python
#Checking accuracy
pred2=log.predict(x_test)
acc2=accuracy_score(pred2,y_test)
acc2
```

```python
#predicting the model with new data set
new=np.array([[5.6,4.0,1.0,0.5],[5.7,2.8,4.0,1.3]])
prediction=log.predict(new)
prediction
```
